1.Import the dataset mobile_prices_2023.csv

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder

price_dataset=pd.read_csv("/content/mobile_prices_2023 - mobile_prices_2023.csv")
price_dataset
```

2.Check the number of null values

```python
price_dataset.columns
```

```python
price_dataset.isnull().sum()
```

3.Drop the null rows and columns

```python
new_dataset=price_dataset.dropna()
new_dataset
```

```python
new_dataset.isnull().sum()
```

4.Apply label encoder for all the string attributes

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
list = ['Phone Name', 'Rating ?/5', 'Number of Ratings', 'RAM', 'ROM/Storage','Back/Rare Camera', 'Front Camera', 'Battery', 'Processor','Price in INR', 'Date of Scraping']
for i in list:
  new_dataset[i]=le.fit_transform(new_dataset[i])
```

```python
new_dataset
```

5.Convert the above attributes into numbers


```python
for i in list:
  new_dataset[i]=new_dataset[i].astype(int)
new_dataset
```

```python
X=new_dataset.drop(["Price in INR"],axis=1)
X
```

```python
y=new_dataset.iloc[:,-2]
y
```

6.Apply MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
scx=mms.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(scx,y,random_state=1,test_size=.2)
```

7. Construct Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(X_train,y_train)
y_pred=mlr.predict(X_test)
y_pred
```

```python
y_test
```

8.Find R2 value

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print('r2 score for perfect model is', r2)
```

9.Apply 6 Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
X=new_dataset.drop('Price in INR',axis=1).values
y=new_dataset[['Price in INR']]
lmr = LinearRegression()
scores = cross_val_score(lmr, X, y, scoring='r2', cv=6)
scores
```

10.Find R2 value

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train,y_test = train_test_split(X, y,train_size = 0.7,test_size = 0.3,random_state = 10)

model = LinearRegression()
mlr.fit(X_train,y_train)
predictions = mlr.predict(X_test)
r2 = r2_score(y_test, predictions)
print('Before Cross Validation: r2 score for perfect model is', r2*100)


scores = cross_val_score(model, X, y, cv=5,scoring='r2')
print("After Cross validation",scores*100)
```
